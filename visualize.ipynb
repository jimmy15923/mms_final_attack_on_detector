```python
for cls in cats:
    images = glob.glob(f'./yolo_results/{cls}/*_raw.png')
    for image in images:
        idx = image.split('_')[-2]
        result_images = glob.glob(f'./yolo_results/{cls}/*{idx}*.png')
        rcnn_images = glob.glob(f'./results/{cls}/*{idx}*.png')
        if len(result_images) == 3 and len(rcnn_images) == 3:
            images = [cv2.resize(cv2.imread(x), (200, 200))[:,:,::-1] for x in sorted(result_images)[::-1]]
            rcnn_images = [cv2.resize(cv2.imread(x), (200, 200))[:,:,::-1] for x in sorted(rcnn_images)[::-1]]
            plt.figure(figsize=(15, 10))
            plt.subplot(231)
            plt.title('Original image')
            plt.imshow(images[0])
            plt.axis('off')
            plt.subplot(232)
            plt.title('FGSM')
            plt.imshow(images[1])
            plt.axis('off')
            plt.subplot(233)
            plt.title('DeepFool')
            plt.imshow(images[2])
            plt.axis('off')
            plt.subplot(234)
            plt.imshow(rcnn_images[0])
            plt.axis('off')
            plt.subplot(235)
            plt.imshow(rcnn_images[1])
            plt.axis('off')
            plt.subplot(236 )
            plt.imshow(rcnn_images[2])
            plt.axis('off')            
            
            plt.savefig(f'qualitative_results/{idx}.png')
            plt.subplots_adjust(wspace=0, hspace=0)
            plt.close()
#     for idx in 
```
